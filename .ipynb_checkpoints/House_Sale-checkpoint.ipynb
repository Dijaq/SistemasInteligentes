## 1. Lectura de datos

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
train = pd.read_csv('all/train.csv')
```

```python
test = pd.read_csv('all/test.csv')
```

```python
#tamaño de los datos de entrenamiento
train.shape
```

```python
#tamaño de los datos de prueba
test.shape
```

```python
#Eliminar columna Id que no brinda informacion importante
train.drop(columns = 'Id', inplace=True)
```

```python
test.drop(columns = 'Id', inplace=True)
```

```python
#mostrar elementos de la data de entrenamiento
train.head()
```

```python
#mostrar elementos de la data de prueba
test.head()
```

```python
#Mostrar cantidad de datos nulos
train.info()
```

## 2. Normalizar el precio de venta

```python
train['SalePrice'] = np.log1p(train['SalePrice'])
```

## 3. Visualización de datos

```python
col = ['SalePrice','OverallQual','GrLivArea', 'GarageCars','TotalBsmtSF','FullBath', 'TotRmsAbvGrd', 'YearBuilt']
sns.set(style='ticks')
sns.pairplot(train[col], height=3, kind='reg')
```

```python
#Visualizar valores nulos
plt.figure(figsize=(15,10))
sns.heatmap(train.isnull())
```

## 4. Eliminando valores nulos

```python
#PoolQC esta columna solo tiene 7 valores no nulos
train['PoolQC'] = train['PoolQC'].fillna('None')
#MiscFeacture  esta columna solo tiene 54 valores no nulos
train['MiscFeature'] = train['MiscFeature'].fillna('None')
#Alley  esta columna solo tiene 91 valores no nulos
train['Alley'] = train['Alley'].fillna('None')
#Fence  esta columna solo tiene 281 valores no nulos
train['Fence'] = train['Fence'].fillna('None')
#FireplaceQu  esta columna solo tiene 770 valores no nulos
train['FireplaceQu'] = train['FireplaceQu'].fillna('None')
```

```python
#Agrupar por el vecino mas cercano y sacar la media de todos los vecinos, son aproximadamente 260 valores reemplazados
train['LotFrontage'] = train.groupby("Neighborhood")['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
#Reemplazar todos los que son tipo objeto por NONE, en todos los casos son menos de 100 valores
for col in ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    train[col] = train[col].fillna('None')
```

```python
#Reemplazar todos los que son tipo float64 por 0, en todos los casos son menos de 100 valores
for col in ['MasVnrArea', 'GarageYrBlt']:
    train[col] = train[col].fillna(int(0))
```

```python
train.info()
```

```python
train.tail()
```

```python
#Visualizar valores nulos
plt.figure(figsize=(15,10))
sns.heatmap(train.isnull())
```

## 5. Codificar string a entero

```python
cols = ('MSZoning','Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood',
    'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
    'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 
    'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType',
        'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition'
       )
```

```python
for c in cols:
    lbl = LabelEncoder()
    lbl.fit(list(train[c].values))
    train[c] = lbl.transform(list(train[c].values))
```

```python
train.head()
```

## 6. Preparando los datos

```python
Y = train['SalePrice']
del train['SalePrice']
```

```python
X = train.values
Y = Y.values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=7)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## 7. Métodos

### Regresion Lineal

```python
from sklearn.linear_model import LinearRegression
algofit = LinearRegression()
algofit.fit(X_train, Y_train)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
algofit = RandomForestRegressor(n_estimators=1000)
algofit.fit(X_train, Y_train)
```

## Predecir

```python
Y_pred = algofit.predict(X_test)
```

## 8. Calculando RMSE de la predicción

```python
print(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
